Dealing with Duplicate values

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("googleplaystore.csv")
data
```

```python

df= data.copy()
```

```python
#checking the total number of rows with duplicate values
df.duplicated().sum()
```

```python
df.duplicated()
```

```python
#This gives me the rows that are having duplicate values
df[df.duplicated()]
```

```python
#Ths gives the number of duplicate values in the particular column(Category)
df.duplicated('Category').sum()
```

```python
df[df.duplicated(keep="first")]    #It will keep the first values and will show me the last value
```

```python
df[df.duplicated(keep="last")]   #It will keep all the last values and will show me the first values
```

```python
df[df.duplicated(keep=False)] #it will show all the values that are duplicate.
```

Dropping the Duplicate Values

```python
#Dropping all the values
df.drop_duplicates(inplace=True)
df
```

```python
df.shape
```

```python
#For Example
cities = ['NY','LA','Chicago','HJouston','Pheonix','LA','Chicago']
Rankings = [1,2,3,4,5,6,7]
Scores = [9.8, 9.5, 9.2, 8.7, 8.5, 8.2, 8.1]
example = {'city': cities, 'Rank':Rankings, 'scores':Scores}

ex = pd.DataFrame(example)
ex
```

```python
# Row 1 (LA, 2, 9.5) is not exactly the same as Row 5 (LA, 6, 8.2).

# Row 2 (Chicago, 3, 9.2) is not exactly the same as Row 6 (Chicago, 7, 8.1).

# Since no entire rows are identical, df.duplicated().sum() returns 0.

ex.duplicated().sum()
```

```python
ex[ex.duplicated()]
```

```python
#this gives the number of cities that are duplicate in the column 'city'
ex.duplicated('city').sum()
```

```python
#Droping the Duplicate cities
ex.drop_duplicates('city', keep ='first', inplace=True)
ex
```
