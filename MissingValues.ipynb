Dealing with Missing Values


```python
#Method 1 : Deleting the columnn with missing data
#method 2 : Deleting the row with missing values
#method 3 : Filling the Missing values (Imputation)
  # (i) Numerical - use mean
  # (ii) Categorical data
#method 4 : Advanced Imputation
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("AB_NYC_2019.csv")
data
```

```python
df = data.copy()
```

```python
df.isnull().sum()
```

1. Deleting the Column with missing value

```python
df.drop('host_name',axis=1, inplace=True)  #this completes deletes the entire column['host_name'] 
```

Deleting the row with missing value

```python
df2 = df.copy()   #copying the df to df2
```

```python
df2.isnull().sum()    
```

```python
df2.dropna(inplace=True)                #all the rows with missing value will be deleted
df2.isna().sum()                  
```

```python
df2.shape
```

Filling the missing value(Imputation)

for Numerical data


```python
df3=data.copy()
df3
```

```python
df3.isna().sum()
```

```python
#this is for imputing the value for single column having numerical data
mean_value = df3['reviews_per_month'].mean()    #we will get the mean of 'reviews_per_month' 
df3['reviews_per_month'].fillna(mean_value, inplace=True)     #replacing the missing value with mean using fillna()
df3
```

Another methos for adding missing values to the column having categorical or number data

```python
#We can also us the Imputer from sci-kit learn for adding the missing values in the columns
from sklearn.impute import SimpleImputer
import numpy as np
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(df.iloc[:,12:13].values)
df3.iloc[:,12:13] = imputer.transform(df.iloc[:,12:13].values)
df3
```

```python
df3.isna().sum()
```

```python
#now handling the catergorical value type data
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer.fit(df3.iloc[:,:].values)
df3.iloc[:,:] = imputer.transform(df3.iloc[:,:].values)
df3
```

```python
df3.isna().sum()
```

```python
df4 = data.copy()
df4
```

4. Advanced Imputation

```python
#This interpolate function will add average(below and above value) and assign it to the missing value.
#this is not ideal for categorical type of data
df4['reviews_per_month'].interpolate(inplace=True)
df4
```
